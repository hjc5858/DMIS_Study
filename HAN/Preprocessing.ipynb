```python
import pandas as pd
import json
import os
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def refine(json_list) :
    sw = stopwords.words('english')
    data_list = list()
    for item in json_list :
        text = [word.lower() for word in item['text'] if (word not in sw) and (word not in string.punctuation)]
        data_list.append(text)
        
    return data_list

        
def file_parser(dir) :
    json_list = list()
    data = open(dir)
    while True:
        l = data.readline()
        if not l :
            break
        json_list.append(json.loads(l))
    data.close()
    data_list = refine(json_list)
        
    return data_list
```

```python
assets = os.listdir('./preprocessed')
print(assets)
```

```python
data_dict = dict()
bag_of_words = list()

for asset in assets :
    asset_dir = './preprocessed/'+asset
    dates = os.listdir(asset_dir)
    date_dict = dict()
    for date in dates :
        data_list = file_parser(asset_dir+'/'+date)
        for i in data_list :
            bag_of_words += i
        date_dict[date] = data_list
    data_dict[asset] = date_dict
    print('appended',asset)

bag_of_words = list(set(bag_of_words))
```
